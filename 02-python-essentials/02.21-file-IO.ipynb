## 写文件

```python
f = open('myfile.txt','w')
f.write('hello world!')
f.close()
```

```python
f = open('myfile.txt','w')
f.write('another hello world!')
f.close()
print(open('myfile.txt').read())
```

```python
f = open('myfile.txt','w+') # 读写模式
f.write('hello world!')
f.seek(6) # 将光标移动到文件的第6个字符处，然后f.read()读出剩下的内容
print(f.read())
f.close()
```

```python
import os
os.remove('myfile.txt')
```

## 关闭文件

```python
f = open('newfile.txt','w')
f.write('hello world')
g = open('newfile.txt','r')
print(repr(g.read()))
```

虽然这里写了内容，但是在关闭之前，内容并没有被写进磁盘。写之后要关闭文件确保内容被保存到磁盘了。

```python
f = open('newfile.txt','w')
f.write('hello world')
f.close()
g = open('newfile.txt','r')
print(repr(g.read()))
```

```python
f = open('newfile.txt','w')
for i in range(3000):
    f.write('hello world'+str(i)+'\n')
    
g = open('newfile.txt','r')
print(g.read())
f.close()
g.close()
```

使用循环写入的内容也不完整。写入完成后要调用close()函数关闭并保存文件。

```python
f = open('newfile.txt','w')
for i in range(3000):
    f.write('hello world'+str(i)+'\n')
f.close()

g = open('newfile.txt','r')
print(g.read())
g.close()
```

```python
import os
os.remove('newfile.txt')
```

出现异常时的读写

```python
f = open('newfile.txt','w')
for i in range(3000):
    x = 1.0/(i-1000)
    f.write('hello world: '+str(i)+'\n')
```

查看已有内容

```python
g = open('newfile.txt','r')
print(g.read())
f.close()
g.close()
```

可以看到，出现异常时，磁盘的写入并没有完成，为此我们可以使用try/except/finally块来关闭文件，这里finally确保关闭文件（不管异常有没有发生，finally都会执行），所有的写入已经完成。

```python
f = open('newfile.txt','w')
try:
    for i in range(3000):
        x = 1.0/(i-1000)
        f.write('hello world: '+str(i)+'\n')
except Exception:
    print('something bad happened')
finally:
    f.close()
```

```python
g = open('newfile.txt','r')
print(g.read())
g.close()
```

## with方法

事实上，python提供了更安全的方法，当with块的内容结束后，python会自动调用它的close方法，确保读写的安全。

```python
with open('newfile.txt','w') as f:
    for i in range(3000):
        x = 1.0/(i-1000)
        f.write('hello world: '+str(i)+'\n')
```

```python
g = open('newfile.txt','r')
print(g.read())
g.close()
```

```python
import os
os.remove('newfile.txt')
```
