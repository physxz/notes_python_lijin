# plot二维图

```python
%matplotlib inline
x = linspace(0,2*pi,50)
plot(sin(x))
```

```python
plot(x,sin(x))
```

```python
plot(x,sin(x),x,sin(2*x))
```

```python
plot(x,sin(x),'r-^')
```

```python
plot(x,sin(x),'b-o',x,sin(2*x),'r-^')
```

与matlab的使用格式基本相同

## 多图

```python
t = linspace(0,2*pi,50)
x = sin(t)
y = cos(t)
figure()
plot(x)
figure()
plot(y)
```

```python
subplot(1,2,1)
plot(x)
subplot(1,2,2)
plot(y)
```

## 向图中添加数据

```python
plot(x)
plot(y)
```

```python
plot(x) # 这个cell有点问题
hold=False
plot(y)
hold=True
```

## 标签

```python
plot(x,label='sin')
plot(y,label='cos')
legend()
```

```python
plot(x)
plot(y)
legend(['sin','cos'])
```

## 坐标轴，标题，网格

```python
plot(x,sin(x))
xlabel('radians')
ylabel('amplitude',fontsize='large')
title('Sin(x)')
```

```python
plot(x,sin(x))
xlabel('radians')
ylabel('amplitude',fontsize='large')
title('Sin(x)')
grid()
```

## 清除，关闭图像

关闭当前图像：`close()`\
关闭所有图像：`close('all')`\
